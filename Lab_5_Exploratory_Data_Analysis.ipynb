## Data Imputation
We want to fill in missing age data instead of just dropping the missing age data rows. One way to do this is by filling in the mean age of all the passengers (imputation).
However we can be smarter about this and check the average age by passenger class. For example:

```python
dataset.isnull().sum()
```

```python
sns.heatmap(dataset.isnull(),cmap='viridis')
```

```python

numerical_columns = dataset.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = dataset.select_dtypes(include=['object']).columns
```

```python
numerical_columns_with_missing_values = numerical_columns[dataset[numerical_columns].isnull().sum() > 0]
print("Numerical Columns with Missing Values =", numerical_columns_with_missing_values)

print("\n")
categorical_columns_with_missing_values = categorical_columns[dataset[categorical_columns].isnull().sum() > 0]
print("Categorical Columns with Missing Values =",categorical_columns_with_missing_values)
```

```python
# Impute missing values for numerical columns with the mean
for col in numerical_columns_with_missing_values:
    dataset[col] = dataset[col].fillna(dataset[col].mean())

# Impute missing values for categorical columns with the mode
for col in categorical_columns_with_missing_values:
    dataset[col] = dataset[col].fillna(dataset[col].mode()[0])

# Check if missing values are handled
print(dataset.isnull().sum())
```
