## Data Imputation
We want to fill in missing age data instead of just dropping the missing age data rows. One way to do this is by filling in the mean age of all the passengers (imputation).
However we can be smarter about this and check the average age by passenger class. For example:

```python
# Get columns with missing values
columns_with_missing_values = train.columns[train.isnull().any()]
print(f"Columns with Missing Values = {columns_with_missing_values.tolist()}")

# From those, select the numeric ones
numeric_cols = train[columns_with_missing_values].select_dtypes(include='number').columns
print("Numeric Columns = ", numeric_cols.tolist())

# Select categorical columns with missing values
categorical_cols = train[columns_with_missing_values].select_dtypes(include='object').columns
print("Categorical Columns = ", categorical_cols.tolist())
```

We can see the wealthier passengers in the higher classes tend to be older, which makes sense. We'll use these average age values to impute based on Pclass for Age.

```python
# Provided manually
numeric_cols = [
    'PhysicalHealthDays', 'MentalHealthDays', 'SleepHours',
    'HeightInMeters', 'WeightInKilograms', 'BMI'
]

categorical_mode_cols = [
    'GeneralHealth', 'LastCheckupTime', 'PhysicalActivities', 'RemovedTeeth',
    'SmokerStatus', 'ECigaretteUsage', 'ChestScan', 'RaceEthnicityCategory',
    'AgeCategory', 'AlcoholDrinkers', 'HIVTesting', 'FluVaxLast12',
    'PneumoVaxEver', 'TetanusLast10Tdap', 'HighRiskLastYear'
]

categorical_ml_cols = [
    'HadHeartAttack', 'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer',
    'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis',
    'HadDiabetes', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty',
    'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing',
    'DifficultyErrands', 'CovidPos'
]
```

```python
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.preprocessing import LabelEncoder
```

### Impute numeric columns with median

```python
# 1. Median imputation for numeric columns
median_imputer = SimpleImputer(strategy='median')
train[numeric_cols] = median_imputer.fit_transform(train[numeric_cols])
```

### Mode imputation for simple categorical columns

```python
# 2. Mode imputation for simple categorical columns
mode_imputer = SimpleImputer(strategy='most_frequent')
train[categorical_mode_cols] = mode_imputer.fit_transform(train[categorical_mode_cols])
```

### Imputation with ML algorithm

```python
# Find rows with missing values before imputation
missing_rows = train[categorical_ml_cols].isnull()

# Prepare the subset for ML-based imputation (only columns that still have missing values)
train_subset = train[categorical_ml_cols].copy()

# 3. Encode only categorical_ml_cols
encoders = {}
for col in categorical_ml_cols:
    le = LabelEncoder()
    train_subset[col] = train_subset[col].astype(str)  # Needed for NaNs
    train_subset[col] = le.fit_transform(train_subset[col])
    encoders[col] = le

# 4. KNN Imputation
knn_imputer = KNNImputer(n_neighbors=5)
train_subset_imputed = pd.DataFrame(
    knn_imputer.fit_transform(train_subset), columns=train_subset.columns
)


# 5. Decode and selectively assign back only to missing entries
for col in categorical_ml_cols:
    # Round and convert to int
    train_subset_imputed[col] = train_subset_imputed[col].round().astype(int)

    # Decode using LabelEncoder
    decoded_col = encoders[col].inverse_transform(train_subset_imputed[col])

    # Assign only to missing values in original train
    train.loc[missing_rows[col], col] = decoded_col[missing_rows[col].values]

# ✅ Final check
print("✅ Imputation complete. Missing values remaining:\n", train.isnull().sum().sum())
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

# --- Binary columns ---
binary_cols = [
    'Sex', 'AlcoholDrinkers', 'HIVTesting', 'FluVaxLast12',
    'PneumoVaxEver', 'TetanusLast10Tdap', 'HighRiskLastYear',
    'HadHeartAttack', 'HadAngina', 'HadStroke', 'HadAsthma', 'HadSkinCancer',
    'HadCOPD', 'HadDepressiveDisorder', 'HadKidneyDisease', 'HadArthritis',
    'HadDiabetes', 'DeafOrHardOfHearing', 'BlindOrVisionDifficulty',
    'DifficultyConcentrating', 'DifficultyWalking', 'DifficultyDressingBathing',
    'DifficultyErrands', 'CovidPos'
]

for col in binary_cols:
    train[col] = train[col].map({'Yes': 1, 'No': 0, 'Female': 0, 'Male': 1})
```

```python
# --- Ordinal columns ---
ordinal_maps = {
    'GeneralHealth': {
        'Poor': 1, 'Fair': 2, 'Good': 3, 'Very good': 4, 'Excellent': 5
    },
    'AgeCategory': {
        'Age 18 to 24': 1, 'Age 25 to 29': 2, 'Age 30 to 34': 3,
        'Age 35 to 39': 4, 'Age 40 to 44': 5, 'Age 45 to 49': 6,
        'Age 50 to 54': 7, 'Age 55 to 59': 8, 'Age 60 to 64': 9,
        'Age 65 to 69': 10, 'Age 70 to 74': 11, 'Age 75 to 79': 12,
        'Age 80 or older': 13
    },
    'LastCheckupTime': {
        'Never': 0,
        '5 or more years ago': 1,
        'Within past 5 years': 2,
        'Within past 2 years': 3,
        'Within past year': 4
    }
}

for col, mapping in ordinal_maps.items():
    train[col] = train[col].map(mapping)
```

```python

# --- One-Hot Encoding ---
onehot_cols = [
    'State', 'RaceEthnicityCategory', 'SmokerStatus',
    'ECigaretteUsage', 'ChestScan', 'RemovedTeeth'
]

train = pd.get_dummies(train, columns=onehot_cols, drop_first=True)
```
